```python
df_1['channel_rate'].value_counts()
```

```python
df_1['pre_sale_rate'].replace('-',np.nan,inplace=True)
df_1_1 = df_1[df_1['pre_sale_rate'].str.contains('%')==1]
df_1_2 = df_1[df_1['pre_sale_rate'].str.contains('%')!=1]
df_1_1['pre_sale_rate'] = df_1_1['pre_sale_rate'].str.strip('%').astype(float) / 100
df_1 = pd.concat([df_1_1,df_1_2])
df_1['pre_sale_rate'] = df_1['pre_sale_rate'].astype(float)
df_1['pre_sale_rate'] = df_1['pre_sale_rate'] * 100
```

```python
df_1['real_sale_rate'].replace('-',np.nan,inplace=True)
df_1_1 = df_1[df_1['real_sale_rate'].str.contains('%')==1]
df_1_2 = df_1[df_1['real_sale_rate'].str.contains('%')!=1]
df_1_1['real_sale_rate'] = df_1_1['real_sale_rate'].str.strip('%').astype(float) / 100
df_1 = pd.concat([df_1_1,df_1_2])
df_1['real_sale_rate'] = df_1['real_sale_rate'].astype(float)
df_1['real_sale_rate'] = df_1['real_sale_rate'] * 100
```

```python
df_1['settle_status'] = df_1['settle_status'].apply(lambda x:1 if x=='已结算' else 0)
df_1.dropna(how='all',inplace=True)
df_1.drop(df_1[df_1['game_id'].isna()].index,inplace=True)
df_1.drop(df_1[df_1['app_channel_id'].isna()].index,inplace=True)
df_1['company_name'] = df_1.apply(lambda x:x.company_name if x.company_name==x.company_name else x.cp_company, axis=1)
df_1.drop_duplicates(subset=['settle_date','game_id','app_channel_id'],keep=False,inplace=True)
```

```python
df_1['settle_date'] = pd.to_datetime(df_1['settle_date']).dt.date
df_mysql_copy['settle_date'] = pd.to_datetime(df_mysql_copy['settle_date']).dt.date
df_1['game_id'] = df_1['game_id'].astype(int)
df_mysql_copy['game_id'] = df_mysql_copy['game_id'].astype(int)
df_1['app_channel_id'] = df_1['app_channel_id'].astype(int)
df_mysql_copy['app_channel_id'] = df_mysql_copy['app_channel_id'].astype(int)
```

```python
df_mysql_copy.set_index(['settle_date','game_id','app_channel_id'],drop=True,inplace=True)
df_1.set_index(['settle_date','game_id','app_channel_id'],drop=True,inplace=True)
```
