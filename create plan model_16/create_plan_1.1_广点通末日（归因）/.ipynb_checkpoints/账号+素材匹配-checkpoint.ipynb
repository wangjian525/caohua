```python
import pandas as pd
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_excel('./投放计划数据明细-所有平台(2022-02-07至2022-02-13).xlsx')
df = df.drop(0)
df = df[['素材', '投放账号', '实际消耗金额', '新增付费创角数', '新创角付费次数', '新创角充值金额']]
df['投放账号'] = df['投放账号'].map(int)
```

```python
mean_roi = df['新创角充值金额'].sum() / df['实际消耗金额'].sum()
mean_roi
```

```python
data = pd.DataFrame({'实际消耗金额':df.groupby(['素材','投放账号'])['实际消耗金额'].sum(),
                         '新增付费创角数':df.groupby(['素材','投放账号'])['新增付费创角数'].sum(),
                        '新创角付费次数':df.groupby(['素材','投放账号'])['新创角付费次数'].sum(),
                         '新创角充值金额':df.groupby(['素材','投放账号'])['新创角充值金额'].sum()}).reset_index()
data = data[data['实际消耗金额']!=0]
data = data[data['新创角充值金额']!=0]
data['roi'] = data['新创角充值金额'] / data['实际消耗金额']
```

```python
data.sort_values('实际消耗金额', ascending=False)
```

## 素材

```python
df_1 = pd.read_excel('./投放计划数据明细-所有平台(2022-02-01至2022-02-13)-素材.xlsx')
df_1 = df_1.drop(0)
df_1 = df_1[['素材', '投放账号', '实际消耗金额', '新增付费创角数', '新创角付费次数', '新创角充值金额']]
df_1['投放账号'] = df_1['投放账号'].map(int)
```

```python
data_1 = pd.DataFrame({'实际消耗金额':df_1.groupby(['素材'])['实际消耗金额'].sum(),
                         '新增付费创角数':df_1.groupby(['素材'])['新增付费创角数'].sum(),
                        '新创角付费次数':df_1.groupby(['素材'])['新创角付费次数'].sum(),
                         '新创角充值金额':df_1.groupby(['素材'])['新创角充值金额'].sum()}).reset_index()
data_1 = data_1[data_1['实际消耗金额']!=0]
data_1 = data_1[data_1['新创角充值金额']!=0]
data_1['roi'] = data_1['新创角充值金额'] / data_1['实际消耗金额']
```

```python
data_1.sort_values('实际消耗金额', ascending=False)
```
