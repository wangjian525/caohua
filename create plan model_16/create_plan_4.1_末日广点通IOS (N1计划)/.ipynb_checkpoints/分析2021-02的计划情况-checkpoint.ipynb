```python
df_now[df_now['label']==1]['deep_bid_type'].value_counts()
```

```python
df_1 = pd.concat([df_now['deep_bid_type'].value_counts(), df_now[df_now['label']==1]['deep_bid_type'].value_counts()], axis=1)
df_1.columns = ['总计划数','首日ROI>=2%的计划数']
df_1['首日达标率'] = df_1['首日ROI>=2%的计划数'] / df_1['总计划数']
```

```python
# 不同优化方式的首日成功率（ROI>=2%）
df_1
```

```python
fig = plt.figure(figsize=(12,8))
ax1 = fig.add_subplot(111)
ax1.bar(df_1.index, df_1['总计划数'].values)
# df_1[['plan_num','label_1_num']].plot.bar()
bar1 = ax1.bar(df_1.index, df_1['总计划数'].values,label = '一部门') 

bar2 = ax1.bar(df_1.index, df_1['首日ROI>=2%的计划数'].values,label = '二部门') 

ax2 = ax1.twinx()        
ax2.plot(df_1.index, df_1['首日达标率'].values, 'r', ms=10, lw=3, marker='o')
plt.show()
```

```python
df_2 = pd.concat([df_now['ad_account_id'].value_counts(), df_now[df_now['label']==1]['ad_account_id'].value_counts()], axis=1)
df_2.columns = ['总计划数','首日ROI>=2%的计划数']
df_2['首日达标率'] = df_2['首日ROI>=2%的计划数'] / df_2['总计划数']
```

```python
df_2 = df_2.fillna(0).sort_values('首日达标率', ascending=False)
```

```python
df_2['总计划数'].mean()
```

```python
df_3 = df_2[df_2['总计划数']>=50]
```

```python
df_3
```

```python
plt.figure(figsize=(20,5),dpi=300)
plt.plot(range(df_3.shape[0]), df_3['首日达标率'].values)
# plt.xticks(range(df_3.shape[0]))
plt.xticks(range(df_3.shape[0]),df_3.index,rotation=45)
plt.show()
```

```python
df_BID_PER_ACTION = df_now[df_now['deep_bid_type'] == 'BID_PER_ACTION']
df_ROI_PACING = df_now[df_now['deep_bid_type'] == 'ROI_PACING']
df_ROI_COEFFICIENT = df_now[df_now['deep_bid_type'] == 'ROI_COEFFICIENT']
df_DEEP_BID_DEFAULT = df_now[df_now['deep_bid_type'] == 'DEEP_BID_DEFAULT']
```

```python
print('每次付费综合首日ROI',round(df_BID_PER_ACTION['new_role_money'].sum() /df_BID_PER_ACTION['amount'].sum(),4))
print('自动ROI综合首日ROI',round(df_ROI_PACING['new_role_money'].sum() /df_ROI_PACING['amount'].sum(),4))
print('ROI系数综合首日ROI',round(df_ROI_COEFFICIENT['new_role_money'].sum() /df_ROI_COEFFICIENT['amount'].sum(),4))
print('默认转化方式综合首日ROI',round(df_DEEP_BID_DEFAULT['new_role_money'].sum() /df_DEEP_BID_DEFAULT['amount'].sum(),4))
```
